# Matrix Class Usage Examples
A from-scratch Python implementation of a Matrix class supporting common linear algebra operations.

```python
# Paste or import your matrix.py here
# from matrix import Matrix, dot_product

class Matrix:

    def __init__(self, arr):
        self.matrix = arr
        for i in range(1, len(arr)):
            if len(arr[0]) != len(arr[i]):
                raise TypeError('Each row must have an equal number of columns')
        self.num_row = len(arr)
        self.num_col = len(arr[0])

    def show(self):
        for row in self.matrix:
            print(row)

    def transpose(self):
        place_holder_matrix = []
        for i in range(self.num_col):
            place_holder_row = []
            for j in range(self.num_row):
                place_holder_row.append(self.matrix[j][i])
            place_holder_matrix.append(place_holder_row)
        self.matrix = place_holder_matrix
        self.num_row, self.num_col = self.num_col, self.num_row
        return self

    def add(self, other_arr):
        if self.num_row != other_arr.num_row or self.num_col != other_arr.num_col:
            raise TypeError('You cannot add/subtract matrices of differing dimensions.')
        for i in range(self.num_row):
            for j in range(self.num_col):
                self.matrix[i][j] += other_arr.matrix[i][j]
        return self

    def subtract(self, other_arr):
        self.add(other_arr.scalar_multiply(-1))
        return self

    def scalar_multiply(self, scalar):
        for i in range(self.num_row):
            for j in range(self.num_col):
                self.matrix[i][j] *= scalar
        return self

    def matrix_multiply(self, other_arr):
        if self.num_col != other_arr.num_row:
            raise TypeError('The number of columns in the first matrix must equal the number of rows in the second!')
        other_arr.transpose()
        resultant_matrix = []
        for i in range(self.num_row):
            row_matrix = []
            for j in range(other_arr.num_col):
                row_matrix.append(dot_product(self.matrix[i], other_arr.matrix[j]))
            resultant_matrix.append(row_matrix)
        other_arr.transpose()
        self.matrix = resultant_matrix
        return self

    def recursive_determinant(self):
        if self.num_col != self.num_row:
            raise TypeError('Rows and Columns must be equal to take the determinant')
        return calculate_determinant(self.matrix)

    def reduced_row_echelon_form(self):
        self.matrix = reduced_row_echelon_form_helper(self.matrix)
        return self

    def inverse_matrix_via_rref(self):
        self.matrix = inverse_matrix_via_rref(self.matrix)
        return self

    def determinant_via_rref(self):
        return determinant_via_rref(self.matrix)


def dot_product(arr1, arr2):
    if len(arr1) != len(arr2):
        raise TypeError('Vectors must be of equal length/dimension')
    return sum(arr1[i] * arr2[i] for i in range(len(arr1)))

def remove_row_and_column(matrix, row, column):
    new_matrix = [r[:] for r in matrix]
    new_matrix.pop(row)
    for r in new_matrix:
        r.pop(column)
    return new_matrix

def calculate_determinant(matrix):
    if len(matrix) == 1:
        return matrix[0][0]
    determinant = 0
    for i in range(len(matrix[0])):
        determinant += (-1)**i * matrix[0][i] * calculate_determinant(remove_row_and_column(matrix, 0, i))
    return determinant

def pivot_exists(matrix, row_index, column_index):
    for i in range(row_index, len(matrix)):
        if matrix[i][column_index] != 0:
            return i
    return -1

def swap_rows(matrix, r1, r2):
    matrix[r1], matrix[r2] = matrix[r2], matrix[r1]

def reduced_row_echelon_form_helper(matrix):
    row_index = 0
    for column_index in range(len(matrix[0])):
        pivot_row_index = pivot_exists(matrix, row_index, column_index)
        if pivot_row_index != -1:
            if pivot_row_index != row_index:
                swap_rows(matrix, pivot_row_index, row_index)
            pivot_number = matrix[row_index][column_index]
            for i in range(len(matrix[row_index])):
                matrix[row_index][i] /= pivot_number
            for row in range(len(matrix)):
                if row == row_index:
                    continue
                scalar = matrix[row][column_index]
                for i in range(len(matrix[row])):
                    matrix[row][i] -= scalar * matrix[row_index][i]
            row_index += 1
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == -0.0:
                matrix[i][j] = 0.0
    return matrix

def inverse_matrix_via_rref(matrix):
    if calculate_determinant(matrix) == 0:
        raise TypeError('This matrix is not invertible, its determinant is 0')
    if len(matrix) != len(matrix[0]):
        raise TypeError('This matrix is not invertible, it is not a square matrix')
    for row_index in range(len(matrix)):
        for column_index in range(len(matrix[row_index])):
            matrix[row_index].append(1 if row_index == column_index else 0)
    rref_matrix = reduced_row_echelon_form_helper(matrix)
    for row_index in range(len(rref_matrix)):
        rref_matrix[row_index] = rref_matrix[row_index][len(rref_matrix[row_index])//2:]
    return rref_matrix

def determinant_via_rref(matrix):
    new_matrix = [row[:] for row in matrix]
    determinant = 1
    row_index = 0
    for column_index in range(len(matrix[0])):
        pivot_row_index = pivot_exists(new_matrix, row_index, column_index)
        if pivot_row_index != -1:
            if pivot_row_index != row_index:
                swap_rows(new_matrix, pivot_row_index, row_index)
                determinant *= -1
            pivot_number = new_matrix[row_index][column_index]
            determinant *= pivot_number
            for i in range(len(new_matrix[row_index])):
                new_matrix[row_index][i] /= pivot_number
            for row in range(len(new_matrix)):
                if row == row_index:
                    continue
                scalar = new_matrix[row][column_index]
                for i in range(len(new_matrix[row])):
                    new_matrix[row][i] -= scalar * new_matrix[row_index][i]
            row_index += 1
        else:
            return 0
    return determinant
```

## 1. Construction & Display

```python
A = Matrix([[1, 2, 3],
            [4, 5, 6],
            [7, 8, 9]])

print('Matrix A:')
A.show()
```

## 2. Transpose

```python
T = Matrix([[1, 2, 3],
            [4, 5, 6]])

print('Before transpose (2x3):')
T.show()

T.transpose()
print('\nAfter transpose (3x2):')
T.show()
```

## 3. Scalar Multiplication

```python
S = Matrix([[1, 2],
            [3, 4]])

S.scalar_multiply(3)
print('After scalar multiply by 3:')
S.show()
```

## 4. Addition & Subtraction

> ⚠️ **Note:** These methods mutate `self` in place. The original matrix is modified after calling `add()` or `subtract()`.

```python
M1 = Matrix([[1, 0],
             [0, 1]])

M2 = Matrix([[2, 3],
             [4, 5]])

M1.add(M2)
print('M1 + M2:')
M1.show()
```

## 5. Matrix Multiplication

```python
# (2x3) @ (3x2) -> (2x2)
P = Matrix([[1, 2, 3],
            [4, 5, 6]])

Q = Matrix([[7,  8],
            [9,  10],
            [11, 12]])

P.matrix_multiply(Q)
print('P @ Q:')
P.show()
```

## 6. Determinant (Recursive Cofactor Expansion)

```python
D = Matrix([[2, 1, 3],
            [0, 4, 1],
            [5, 2, 0]])

print('Determinant (recursive):', D.recursive_determinant())
```

## 7. Determinant (via RREF)

More numerically efficient for large matrices — O(n³) vs O(n!) for cofactor expansion.

```python
D2 = Matrix([[2, 1, 3],
             [0, 4, 1],
             [5, 2, 0]])

print('Determinant (via RREF):', D2.determinant_via_rref())
```

## 8. Reduced Row Echelon Form (RREF)

```python
# System: x + 2y = 5, 3x + 4y = 6  -> augmented matrix
R = Matrix([[1, 2, 5],
            [3, 4, 6]])

print('Augmented matrix [A|b]:')
R.show()

R.reduced_row_echelon_form()
print('\nRREF (solution: last column):')
R.show()
```

## 9. Matrix Inverse (via RREF)

```python
INV = Matrix([[4, 7],
              [2, 6]])

print('Original:')
INV.show()

INV.inverse_matrix_via_rref()
print('\nInverse:')
INV.show()
```

## 10. Error Handling

```python
# Singular matrix (det = 0) -> not invertible
try:
    Singular = Matrix([[1, 2],
                       [2, 4]])
    Singular.inverse_matrix_via_rref()
except TypeError as e:
    print('Caught expected error:', e)

# Dimension mismatch on addition
try:
    X = Matrix([[1, 2, 3]])
    Y = Matrix([[1, 2]])
    X.add(Y)
except TypeError as e:
    print('Caught expected error:', e)
```
